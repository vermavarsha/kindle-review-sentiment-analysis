```python
#positive review 1 and negative 0
df['rating']=df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
#1.lower all text
df['reviewText']=df['reviewText'].str.lower()
```

```python
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python
##remove special character
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+','',x))
##remove stopword
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url
df['reviewText']=df['reviewText'].apply(lambda x:re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Remove html tags
df['reviewText']=df['reviewText'].apply(lambda x:BeautifulSoup(x,'lxml').get_text())
##remove aditional spaces
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
nltk.download('wordnet')
```

```python
def lemmatize_words(text):
  return " ".join(lemmatizer.lemmatize(word) for word in text.split())
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
df
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
X_train_tfidf
```

```python
from sklearn.naive_bayes import GaussianNB
model_bow=GaussianNB()
model_bow.fit(X_train_bow,y_train)

model_tfidf=GaussianNB()
model_tfidf.fit(X_train_tfidf,y_train)
```

```python
y_pred_bow=model_bow.predict(X_test_bow)
y_pred_tfidf=model_tfidf.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("BOW")
print(accuracy_score(y_test,y_pred_bow))
print(confusion_matrix(y_test,y_pred_bow))
print(classification_report(y_test,y_pred_bow))
print(" ")
print("TFIDF")
print(accuracy_score(y_test,y_pred_tfidf))
print(confusion_matrix(y_test,y_pred_tfidf))
print(classification_report(y_test,y_pred_tfidf))
```

```python
import joblib
import numpy as np
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Load models
xgb_model = joblib.load('xgboost_model_smote.pkl')
w2v_model = Word2Vec.load('word2vec_model.model')

# Your avg_word2vec function
def avg_word2vec(doc, w2v_model):
    vectors = [w2v_model.wv[word] for word in doc if word in w2v_model.wv.key_to_index]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(w2v_model.vector_size)

# === TEST INPUT ===
text = "This book was truly inspiring and beautifully written!"

# Preprocess
tokens = word_tokenize(text.lower())
vector = avg_word2vec(tokens, w2v_model).reshape(1, -1)

# Predict
prediction = xgb_model.predict(vector)[0]
print("Prediction:", "Positive" if prediction == 1 else "Negative")
```
